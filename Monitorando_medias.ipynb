```python
import yfinance as yf
import matplotlib.pyplot as plt

# Função para calcular e plotar a ação, média móvel de 90 dias e desvio de 2 sigmas
def plot_bbas3_with_moving_average_and_deviation():
    # Obter dados históricos desde 2008
    ticker = 'BBAS3.SA'
    data = yf.Ticker(ticker).history(start='2022-01-01')

    # Calcular a média móvel de 90 dias e o desvio padrão
    data['MA90'] = data['Close'].rolling(window=90).mean()
    data['StdDev'] = data['Close'].rolling(window=90).std()
    data['UpperBand'] = data['MA90'] + 2 * data['StdDev']
    data['LowerBand'] = data['MA90'] - 2 * data['StdDev']

    # Plotar os dados
    plt.figure(figsize=(14, 7))

    plt.plot(data.index, data['Close'], label='BBAS3', color='blue')
    plt.plot(data.index, data['MA90'], label='Média Móvel de 90 dias', color='orange')
    plt.plot(data.index, data['UpperBand'], label='2 Sigma Superior', color='green', linestyle='--')
    plt.plot(data.index, data['LowerBand'], label='2 Sigma Inferior', color='red', linestyle='--')

    # Configurações do gráfico
    plt.title('BBAS3 com Média Móvel de 90 Dias e Desvio de 2 Sigmas')
    plt.xlabel('Data')
    plt.ylabel('Preço (R$)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Chamar a função para plotar o gráfico
plot_bbas3_with_moving_average_and_deviation()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Função para calcular e plotar a ação, média móvel de 90 dias, desvio de 2 sigmas e pontos de compra
def plot_bbas3_with_moving_average_and_deviation():
    # Obter dados históricos desde 2008
    ticker = 'BBAS3.SA'
    data = yf.Ticker(ticker).history(start='2022-01-01')

    # Calcular a média móvel de 90 dias e o desvio padrão
    data['MA90'] = data['Close'].rolling(window=90).mean()
    data['StdDev'] = data['Close'].rolling(window=90).std()
    data['UpperBand'] = data['MA90'] + 2 * data['StdDev']
    data['LowerBand'] = data['MA90'] - 2 * data['StdDev']

    # Encontrar os pontos de compra
    buy_signals = data[(data['Close'] > data['MA90'] + 2 * data['StdDev'])]

    # Plotar os dados
    plt.figure(figsize=(14, 7))

    plt.plot(data.index, data['Close'], label='BBAS3', color='blue')
    plt.plot(data.index, data['MA90'], label='Média Móvel de 90 dias', color='orange')
    plt.plot(data.index, data['UpperBand'], label='2 Sigma Superior', color='green', linestyle='--')
    plt.plot(data.index, data['LowerBand'], label='2 Sigma Inferior', color='red', linestyle='--')
    plt.scatter(buy_signals.index, buy_signals['Close'], label='Pontos de Compra', color='purple', marker='^', alpha=1)

    # Configurações do gráfico
    plt.title('BBAS3 com Média Móvel de 90 Dias, Desvio de 2 Sigmas e Pontos de Compra')
    plt.xlabel('Data')
    plt.ylabel('Preço (R$)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Chamar a função para plotar o gráfico
plot_bbas3_with_moving_average_and_deviation()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Função para calcular e plotar a ação, média móvel de 90 dias, desvio de 2 sigmas e pontos de compra
def plot_bbas3_with_moving_average_and_deviation():
    # Obter dados históricos desde 2008
    ticker = 'BBAS3.SA'
    data = yf.Ticker(ticker).history(start='2020-01-01')

    # Calcular a média móvel de 90 dias e o desvio padrão
    data['MA90'] = data['Close'].rolling(window=90).mean()
    data['StdDev'] = data['Close'].rolling(window=90).std()
    data['UpperBand'] = data['MA90'] + 2 * data['StdDev']
    data['LowerBand'] = data['MA90'] - 2 * data['StdDev']

    # Encontrar os pontos de compra (quando o preço está 2 sigmas abaixo da média móvel)
    buy_signals = data[(data['Close'] < data['LowerBand'])]

    # Plotar os dados
#     plt.figure(figsize=(14, 7))

    plt.plot(data.index, data['Close'], label='BBAS3', color='blue')
    plt.plot(data.index, data['MA90'], label='Média Móvel de 90 dias', color='orange')
    plt.plot(data.index, data['UpperBand'], label='2 Sigma Superior', color='green', linestyle='--')
    plt.plot(data.index, data['LowerBand'], label='2 Sigma Inferior', color='red', linestyle='--')
    plt.scatter(buy_signals.index, buy_signals['Close'], label='Pontos de Compra', color='purple', marker='^', alpha=1)

    # Configurações do gráfico
#     plt.title('BBAS3 com Média Móvel de 90 Dias, Desvio de 2 Sigmas e Pontos de Compra')
    plt.xlabel('Data')
    plt.ylabel('Preço (R$)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('bbas3_90dias_ma.png')
    plt.show()
    

# Chamar a função para plotar o gráfico
plot_bbas3_with_moving_average_and_deviation()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Função para calcular e plotar o preço do BTC e o Índice Mayer ajustado para 90 dias
def plot_btc_with_mayer_index():
    # Obter dados históricos do Bitcoin
    ticker = 'BTC-USD'
    data = yf.Ticker(ticker).history(period='max')

    # Calcular a média móvel de 90 dias
    data['MA90'] = data['Close'].rolling(window=90).mean()

    # Calcular o Índice Mayer (preço / média móvel de 90 dias)
    data['Mayer Index'] = data['Close'] / data['MA90']

    # Plotar os dados
#     plt.figure(figsize=(14, 7))

   # plt.plot(data.index, data['Close'], label='Preço do BTC', color='blue')
    plt.plot(data.index, data['Mayer Index'], label='Índice Mayer (90 dias)', color='orange')
    plt.axhline(y=1, color='r', linestyle='-')

    # Configurações do gráfico
#     plt.title('Preço do BTC com Índice Mayer Ajustado para 90 Dias')
    plt.xlabel('Data')
    plt.ylabel('Valor')
    plt.legend()
    plt.grid(True)
    plt.show()

# Chamar a função para plotar o gráfico
plot_btc_with_mayer_index()
```
