```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Baixar os dados históricos do Ibovespa
ibovespa = yf.Ticker('^BVSP')
historical_data = ibovespa.history(period='max')

# Calcular a média móvel de 90 dias e o desvio padrão
historical_data['MA90'] = historical_data['Close'].rolling(window=90).mean()
historical_data['STD90'] = historical_data['Close'].rolling(window=90).std()

# Inicializar variáveis
initial_cash = 1000
cash = initial_cash
shares = 0

# Iterar sobre os dados históricos e aplicar a estratégia
for i in range(len(historical_data)):
    if i >= 90:  # Para garantir que temos pelo menos 90 dias de dados
        price = historical_data['Close'][i]
        ma90 = historical_data['MA90'][i]
        std90 = historical_data['STD90'][i]
        
        # Comprar se a diferença for maior que 2 desvios padrão positivos
        if (price - ma90) > 2 * std90 and cash > 0:
            shares = cash / price
            cash = 0
            print(f'Comprou {shares:.4f} ações a {price:.2f} em {historical_data.index[i]}')
        
        # Vender se a diferença for maior que 2 desvios padrão negativos
        elif (price - ma90) < -2 * std90 and shares > 0:
            cash = shares * price
            shares = 0
            print(f'Vendeu ações a {price:.2f} em {historical_data.index[i]}, cash = {cash:.2f}')

# Calcular o valor final
final_value = cash if shares == 0 else shares * historical_data['Close'][-1]

# Plotar os dados históricos do Ibovespa com média móvel de 90 dias
plt.figure(figsize=(14, 7))
plt.plot(historical_data['Close'], label='Ibovespa')
plt.plot(historical_data['MA90'], label='Média Móvel 90 Dias', linestyle='--')

# Configurações do gráfico
plt.title('Ibovespa com Média Móvel de 90 Dias')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()

print(f'Valor final do investimento: R${final_value:.2f}')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Baixar os dados históricos do BBAS3
bbas3 = yf.Ticker('BBAS3.SA')
historical_data = bbas3.history(period='max')

# Calcular a média móvel de 90 dias e o desvio padrão
historical_data['MA90'] = historical_data['Close'].rolling(window=90).mean()
historical_data['STD90'] = historical_data['Close'].rolling(window=90).std()

# Inicializar variáveis
initial_cash = 1000

# Simulação sem taxa
cash_no_tax = initial_cash
shares_no_tax = 0

# Simulação com taxa
cash_with_tax = initial_cash
shares_with_tax = 0

# Iterar sobre os dados históricos e aplicar a estratégia
for i in range(len(historical_data)):
    if i >= 90:  # Para garantir que temos pelo menos 90 dias de dados
        price = historical_data['Close'][i]
        ma90 = historical_data['MA90'][i]
        std90 = historical_data['STD90'][i]
        
        # Estratégia sem taxa
        if (price - ma90) > 2 * std90 and cash_no_tax > 0:
            shares_no_tax = cash_no_tax / price
            cash_no_tax = 0
        elif (price - ma90) < -2 * std90 and shares_no_tax > 0:
            cash_no_tax = shares_no_tax * price
            shares_no_tax = 0

        # Estratégia com taxa
        if (price - ma90) > 2 * std90 and cash_with_tax > 0:
            shares_with_tax = cash_with_tax / price
            cash_with_tax = 0
        elif (price - ma90) < -2 * std90 and shares_with_tax > 0:
            sale_value = shares_with_tax * price
            cost_basis = initial_cash / shares_with_tax
            profit = sale_value - cost_basis
            tax = 0.15 * profit
            cash_with_tax = sale_value - tax
            shares_with_tax = 0

# Calcular o valor final
final_value_no_tax = cash_no_tax if shares_no_tax == 0 else shares_no_tax * historical_data['Close'][-1]
final_value_with_tax = cash_with_tax if shares_with_tax == 0 else shares_with_tax * historical_data['Close'][-1]

# Plotar os dados históricos do BBAS3 com média móvel de 90 dias
plt.figure(figsize=(14, 7))
plt.plot(historical_data['Close'], label='BBAS3')
plt.plot(historical_data['MA90'], label='Média Móvel 90 Dias', linestyle='--')

# Configurações do gráfico
plt.title('BBAS3 com Média Móvel de 90 Dias')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()

print(f'Valor final do investimento sem taxa: R${final_value_no_tax:.2f}')
print(f'Valor final do investimento com taxa: R${final_value_with_tax:.2f}')
print(f'Diferença: R${final_value_no_tax - final_value_with_tax:.2f}')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Baixar os dados históricos do BBAS3
bbas3 = yf.Ticker('BBAS3.SA')
historical_data = bbas3.history(period='max',start='2004-01-01')
dividends = bbas3.dividends

# Calcular a média móvel de 90 dias e o desvio padrão
historical_data['MA90'] = historical_data['Close'].rolling(window=90).mean()
historical_data['STD90'] = historical_data['Close'].rolling(window=90).std()

# Inicializar variáveis
initial_cash = 1000
cash = initial_cash
shares = 0

# Iterar sobre os dados históricos e aplicar a estratégia
for i in range(len(historical_data)):
    if i >= 90:  # Para garantir que temos pelo menos 90 dias de dados
        price = historical_data['Close'][i]
        ma90 = historical_data['MA90'][i]
        std90 = historical_data['STD90'][i]
        
        # Comprar ações inicialmente
        if shares == 0 and cash > 0:
            shares = cash / price
            cash = 0
            print(f'Comprou {shares:.4f} ações a {price:.2f} em {historical_data.index[i]}')

        # Reinvestir dividendos quando a diferença for maior que 2 desvios padrão positivos
        if (price - ma90) > 2 * std90 and cash > 0:
            new_shares = cash / price
            shares += new_shares
            cash = 0
            print(f'Reinvestiu dividendos e comprou {new_shares:.4f} ações a {price:.2f} em {historical_data.index[i]}')
        
        # Verificar se há dividendos a serem reinvestidos
        if historical_data.index[i] in dividends.index:
            dividend_amount = dividends[historical_data.index[i]] * shares
            cash += dividend_amount
            print(f'Recebeu dividendos de R${dividend_amount:.2f} em {historical_data.index[i]}')

# Calcular o valor final
final_value = shares * historical_data['Close'][-1]

# Plotar os dados históricos do BBAS3 com média móvel de 90 dias
plt.figure(figsize=(14, 7))
plt.plot(historical_data['Close'], label='BBAS3')
plt.plot(historical_data['MA90'], label='Média Móvel 90 Dias', linestyle='--')

# Configurações do gráfico
plt.title('BBAS3 com Média Móvel de 90 Dias')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()

print(f'Valor final do investimento: R${final_value:.2f}')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
from datetime import timedelta

# Baixar os dados históricos do BBAS3
bbas3 = yf.Ticker('BBAS3.SA')
historical_data = bbas3.history(period='max')
dividends = bbas3.dividends

# Calcular a média móvel de 90 dias e o desvio padrão
historical_data['MA90'] = historical_data['Close'].rolling(window=90).mean()
historical_data['STD90'] = historical_data['Close'].rolling(window=90).std()

# Inicializar variáveis
initial_cash = 1000
cash = initial_cash
shares = 0
pending_dividends = {}

# Iterar sobre os dados históricos e aplicar a estratégia
for i in range(len(historical_data)):
    if i >= 90:  # Para garantir que temos pelo menos 90 dias de dados
        date = historical_data.index[i]
        price = historical_data['Close'][i]
        
        # Comprar ações inicialmente
        if shares == 0 and cash > 0:
            shares = cash / price
            cash = 0
            print(f'Comprou {shares:.4f} ações a {price:.2f} em {date}')
        
        # Verificar se há dividendos a serem recebidos
        if date in dividends.index:
            dividend_amount = dividends[date] * shares
            pending_dividends[date + timedelta(days=30)] = dividend_amount
            print(f'Recebeu dividendos de R${dividend_amount:.2f} em {date}, a serem aplicados em {date + timedelta(days=30)}')

        # Reinvestir dividendos após um mês
        if date in pending_dividends:
            cash += pending_dividends.pop(date)
            if cash > 0:
                new_shares = cash / price
                shares += new_shares
                cash = 0
                print(f'Reinvestiu dividendos e comprou {new_shares:.4f} ações a {price:.2f} em {date}')

# Calcular o valor final
final_value = shares * historical_data['Close'][-1]

# Plotar os dados históricos do BBAS3 com média móvel de 90 dias
plt.figure(figsize=(14, 7))
plt.plot(historical_data['Close'], label='BBAS3')
plt.plot(historical_data['MA90'], label='Média Móvel 90 Dias', linestyle='--')

# Configurações do gráfico
plt.title('BBAS3 com Média Móvel de 90 Dias')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()

print(f'Valor final do investimento: R${final_value:.2f}')
```
